# Data Cleaning

### Load Raw Dataset

We begin by loading the raw credit default dataset from an Excel file. The dataset includes 30,000 records and 25 columns, with features ranging from demographic information to payment history and bill amounts.

```python
import pandas as pd

df = pd.read_excel('../data/credit-default-project.xls')

df.head()
```

### Remove Index and ID Columns

We remove the index row and the ID column to eliminate identifiers to avoid introducing bias or noise.


```python
df = df.drop(index=0).reset_index(drop=True)
df = df.drop(df.columns[0], axis=1)
df.head()
```

### Inspect Dataset Structure

We confirm the dataset contains 30,000 rows and 24 columns. At this point, all columns are of object dtype and need conversion to numeric types.

```python
df.shape
```

```python
df.info()
```

### Convert Features to Numeric

All columns are explicitly converted to numeric format. Any non-numeric values are coerced to `NaN`. This step is essential for downstream statistical analysis and model training.


```python
df = df.apply(pd.to_numeric, errors='coerce')
```

```python
df.info()
```

### Null Value Check

We verify that no null values remain in the dataset after conversion. This confirms successful parsing and cleaning of initial values.

```python
df.isnull().sum()
```

### Scan for Suspicious Placeholder Values

We define a set of suspicious placeholders often used to represent missing or invalid data. These include numeric stand-ins (e.g., -1, 9999) and string values (e.g., 'NA', 'null'). We then scan the dataset column by column to identify and count such entries.


```python
cols_to_check = [
                 'X6', 'X7', 'X8','X9', 'X10', 'X11',
                 ]

# Should only contain -2 through 9
placeholder_values = [ -99, -999, 999, 9999, 99999, 'NA', 'N/A', '?', 'null', 'NULL']


for col in cols_to_check:

    col_as_string = df[col].astype(str)
    
    mask = col_as_string.isin(map(str, placeholder_values))
    sus_rows = df[mask]
    
    if not sus_rows.empty:
        print(f"\n\n Suspicious placeholder values found in column {col}\n\n")
        print(sus_rows[[col]])
    else:
        print(f"\nNo suspicious values found in column {col}\n")
```

```python
cols_to_check = [
                 'X12', 'X13', 'X14','X15', 'X16', 'X17',
                 ]
# Positive and Negative values are acceptable so we remove them from placeholder search.
placeholder_values = [
    'NA', 
    'N/A', 
    '?', 
    'null', 
    'NULL'
    ]

for col in cols_to_check:

    col_as_string = df[col].astype(str)
    
    mask = col_as_string.isin(map(str, placeholder_values))
    sus_rows = df[mask]
    
    if not sus_rows.empty:
        print(f"\n\nSuspicious placeholder values found in column {col}\n\n")
        print(sus_rows[[col]]) 
    else:
        print(f"\nNo suspicious values found in column {col}\n")
```

```python

cols_to_check = [
                 'X18', 'X20', 'X21', 'X22', 'X23', 
                 ]

# Positive values are acceptable so we remove them from the search. 
placeholder_values = [
    -1,
    -99, 
    -999, 
    -9999,
    'NA', 
    'N/A', 
    '?', 
    'null', 
    'NULL'
    ]

for col in cols_to_check:

    col_as_string = df[col].astype(str)
    
    mask = col_as_string.isin(map(str, placeholder_values))
    sus_rows = df[mask]
    
    if not sus_rows.empty:
        print(f"\n\nSuspicious placeholder values found in column {col}\n\n")
        print(sus_rows[[col]].head(20))  # show first 20 suspicious values
    else:
        print(f"\nNo suspicious values found in column {col}\n")
```

### Visually Inspect unique value ranges for each feature.

We check for abnormal data ranges for each feature.

```python
for feature in df.columns:
    print('\n\n-------------------------------------------------------------')
    print(f'{feature} \nUnique Values: {df[feature].unique()}\n')
    print('-------------------------------------------------------------\n\n')
```

### Manage Duplicates

We look for fully identical duplicate rows and remove them from the dataset. 

```python
dupe_flags = df.duplicated(keep=False)
dupes = df[dupe_flags].copy()

dupes["duplicate_group"] = dupes.groupby(list(df.columns)).ngroup()

dupes_sorted = dupes.sort_values(by="duplicate_group")

print(f"Total duplicate groups: {dupes_sorted['duplicate_group'].nunique()}")
print(f"Total rows involved in duplicates: {dupes_sorted.shape[0]}")
```

We can visually inspect the duplicates to verify. 

```python
import pandas as pd
pd.set_option("display.max_rows", None)
pd.set_option("display.max_columns", None) 
dupes_sorted
```

```python
df_no_dupes = df.drop_duplicates()

print(f"Original rows: {df.shape[0]}")
print(f"After removing duplicates: {df_no_dupes.shape[0]}")
print(f"Removed {df.shape[0] - df_no_dupes.shape[0]} duplicate rows")
```

```python
remaining_dupes = df_no_dupes[df_no_dupes.duplicated(subset=df_no_dupes.columns, keep=False)]

print(f"Number of fully identical duplicate rows: {remaining_dupes.shape[0]}")
```

```python
df_no_dupes.to_csv('../data/clean.csv', index=False)
```

## Summary of Data Cleaning

In this notebook, we completed a comprehensive data cleaning pipeline for the credit default dataset. Key steps and findings include:

- **Loaded** the original dataset and removed metadata fields (`index`, `ID`).
- **Converted** all features to numeric types, coercing non-numeric values safely.
- **Validated** the dataset had no missing (`NaN`) values post-cleaning.
- **Scanned for placeholder values** (e.g., `-1`, `9999`) and identified them in both billing and payment columns.
- **Flagged negative or zero balances**, particularly in bill and payment history fields.
- **Audited categorical features** for unexpected values (e.g., 7 unique levels in education, 4 in marital status).
- **Identified and removed** fully duplicated rows (70 duplicates across 35 identical groups).
- **Exported** the cleaned dataset as `clean.csv` for downstream processing.

The dataset is now numerically consistent, stripped of placeholder artifacts, and free of duplicates  ready for feature engineering and modeling.
